```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\SAMSUNG\Documents\메가스터디KDT\팀프로젝트2차(데이터)\ott_blog,cafe_results (raw data).csv"
)

print(df.head())
```

```python
df['text'] = df['제목'].fillna('') + ' ' + df['설명'].fillna('')
```

```python
print(df[['제목', '설명', 'text']].head())
```

```python
# 포함 키워드
include_keywords = [
    "OTT",
    "넷플릭스",
    "드라마",
    "정주행",
    "콘텐츠",
    "시리즈",
    "러닝타임",
    "쇼츠",
    "숏폼",
    "하이라이트",
    "요약",
    "지루",
    "호흡이",
    "볼 게",
    "볼게"
]
```

```python
#필터링

df_filtered = df[
    df['text'].apply(
        lambda x: any(keyword in x for keyword in include_keywords)
    )
]
```

```python
#결과 확인
print(df_filtered[['제목', '설명']].head(10))
```

```python
# 제외 키워드 만들기

exclude_keywords = [
    "VPN",
    "자동차",
    "출퇴근",
    "배송",
    "여행",
    "헬스",
    "공부",
    "맛집"
]
```

```python
# 노이즈 제거

df_filtered = df_filtered[
    ~df_filtered['text'].apply(
        lambda x: any(keyword in x for keyword in exclude_keywords)
    )
]
```

```python
print(df_filtered[['제목', '설명']].head(20))
```

```python
# 텍스트 하나로 합치기

text_data = ' '.join(df_filtered['text'].astype(str))
```

```python
# 단어 추출

words = text_data.split()
```

```python
# Counter 사용

from collections import Counter

word_counts = Counter(words)
```

```python
# 많이 나온 단어 보기

print(word_counts.most_common(100))
```

```python
from collections import Counter

co_occurrence = Counter()

target_word = "숏폼"

for text in df_filtered['text']:

    words = text.split()

    if target_word in words:

        for word in words:
            if word != target_word:
                co_occurrence[word] += 1

print(co_occurrence.most_common(50))
```
