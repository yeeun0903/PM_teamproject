```python
import requests
import pandas as pd
import time

# 네이버 API 정보 입력
client_id = "876egX596tQh5hGq7lGp"
client_secret = "WeHJUOQ5V7"

# 검색할 키워드 리스트
queries = [
    "OTT 너무 길다",
    "드라마 호흡 길다",
    "정주행 힘들다",
    "드라마 요약만 본다",
    "하이라이트만 본다",
    "짧은 콘텐츠 선호",
    "5분 드라마",
    "쇼츠로 본다",
    "볼 거 찾기 힘들다",
    "OTT 지루하다",
    "드라마 늘어진다",
    "요약 영상",
    "숏드라마",
    "릴스처럼",
    "틱톡 같다",
    "짧은 영상 대세",
    "숏폼 대세"
]

# 결과 저장용 리스트
all_results = []

# API URL
url = "https://openapi.naver.com/v1/search/blog.json"

# 헤더 설정
headers = {
    "X-Naver-Client-Id": client_id,
    "X-Naver-Client-Secret": client_secret
}

# 각 키워드별 검색 실행
for query in queries:

    params = {
        "query": query,
        "display": 100,   # 검색 결과 개수
        "sort": "sim"    # 정확도순
    }

    response = requests.get(url, headers=headers, params=params)

    print(f"{query} → 상태코드: {response.status_code}")

    if response.status_code == 200:

        data = response.json()

        for item in data['items']:

            result = {
                "검색키워드": query,
                "제목": item['title'],
                "설명": item['description'],
                "링크": item['link'],
                "작성일": item['postdate'],
                "블로그명": item['bloggername']
            }

            all_results.append(result)

    else:
        print("에러 발생:", response.text)

    # 너무 빠른 요청 방지
    time.sleep(1)

# 데이터프레임 변환
df = pd.DataFrame(all_results)

# HTML 태그 제거
df['제목'] = df['제목'].str.replace(r'<.*?>', '', regex=True)
df['설명'] = df['설명'].str.replace(r'<.*?>', '', regex=True)

# CSV 저장
df.to_csv("ott_blog_results.csv", index=False, encoding="utf-8-sig")

print("저장 완료!")
print(df.head())
```
